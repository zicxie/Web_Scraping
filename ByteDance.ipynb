```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
timeout = 5

browser = webdriver.Chrome()
browser.get(new_url)
element_present = EC.presence_of_element_located((By.CLASS_NAME, 'listItems__1q9i5'))
WebDriverWait(browser, timeout).until(element_present)
browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
#browser.implicitly_wait(10)
soup = BeautifulSoup(browser.page_source,'html.parser')
Listings = soup.find_all('div',class_='listItems__1q9i5')[0]
```

```python
Listings
```

```python
links = []
for i in range(int(max_num)):
    add = Listings.find_all('a')[i]['href']
    link = "https://job.bytedance.com"+str(add)
    links.append(link)
```

## Main Scraping

```python
content = []

driver= webdriver.Chrome()
cnt = 1317
for j in links[1800:2000]:
    driver.execute_script("window.open('');")
    print(cnt)
    cnt+=1

    driver.switch_to.window(driver.window_handles[1])

    driver.get(j)
    sleep(2)
 #   soup = BeautifulSoup(driver.page_source,'html.parser')
    ### Title
    try:
        title = driver.find_element(By.XPATH,'//*[@id="bd"]/section/section/main/div/div/div[1]').text
    except:
        title = 'NA'
    #   title = soup.find('div',class_="job-header sofiaBold").get_text()
    ### Location
    try:
        city = driver.find_element(By.XPATH,'//*[@id="bd"]/section/section/main/div/div/div[2]/span[1]').text
    except:
        city = 'NA'
    ### Function
    try:
        function = driver.find_element(By.XPATH,'//*[@id="bd"]/section/section/main/div/div/div[2]/span[2]').text
    except:
        function = 'NA'
    ### jd
    try:
        jd = driver.find_element(By.XPATH,'//*[@id="bd"]/section/section/main/div/div/div[4]').text
    except:
        jd = 'NA'
    ### jr
    try:
        jr = driver.find_element(By.XPATH,'//*[@id="bd"]/section/section/main/div/div/div[6]').text
    except:
        jr = 'NA'
    # close the active tab
    driver.close()
    # Switch back to the first tab
    driver.switch_to.window(driver.window_handles[0])

    content.append([title,city,function,jd,jr,j])

#driver.get(start_url)
#soup = BeautifulSoup(driver.page_source,'html.parser')
#soup.find_all('div',class_="job-header sofiaBold")[0].get_text()
```
