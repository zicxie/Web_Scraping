```python
df_sum = []
index = 1
for i in df['LinkedIn']:
    try:
        driver.get(i+'/jobs/')
        element_click = EC.element_to_be_clickable((By.LINK_TEXT, 'Insights'))
        WebDriverWait(driver, 5).until(element_click)
        menu = driver.find_element_by_class_name('org-grid__left-rail')
        try:
            employees = driver.find_element_by_partial_link_text('See all').text
        except:
            employees =''

        try:
            element_present = EC.presence_of_element_located((By.CLASS_NAME, 'org-jobs-job-search-form-module__headline'))
            WebDriverWait(driver, 3).until(element_present)
            jobopenings = int(driver.find_element_by_class_name('org-jobs-job-search-form-module__headline').text.split('job')[0].strip().split(' ')[-1])
        except:
            jobopenings = ''

        try:
            menu.find_element_by_link_text('Insights').click()
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        except:
            print('No.'+str(index)+' Insights Page Not Clickable')
        #sleep(random.random()*10+3)
        try:
            element_present = EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'td.org-insights-module__summary-block'))
            WebDriverWait(driver, 7).until(element_present)
            yr_sum = driver.find_elements_by_css_selector('td.org-insights-module__summary-block')
            hlfyr=yr_sum[1].text
            oneyr=yr_sum[2].text
            twoyr=yr_sum[3].text
        except:
            hlfyr=''
            oneyr=''
            twoyr=''
            print('No.'+str(index)+' Insights Page Missing Info')
        summary = [jobopenings,employees,hlfyr,oneyr,twoyr,i]
    except:
        summary = ['na','na','na','na','na',i]
        print(str(index)+' Link Not working')
    df_sum.append(summary)
    print(str(index)+ ' Finished')
    index+=1
df_final = pd.DataFrame(df_sum,columns=['job_openings','# employees','hlfyr','oneyr','twoyr','link'])
df_final
```

```python
def scroll_to_bottom(driver):

    old_position = 0
    new_position = None

    while new_position != old_position:
        # Get old scroll position
        old_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
        # Sleep and Scroll
        time.sleep(1)
        driver.execute_script((
                "var scrollingElement = (document.scrollingElement ||"
                " document.body);scrollingElement.scrollTop ="
                " scrollingElement.scrollHeight;"))
        # Get new position
        new_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
```
